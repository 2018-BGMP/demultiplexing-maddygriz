```python
import argparse
import numpy as np
import math
import statistics
import matplotlib
import matplotlib.pyplot as plt

file1 = "R1.txt"
file2 = "R2.txt"
file3 = "R3.txt"
file4 = "R4.txt"
index = 8
read = 101
```

```python
indexArray = np.zeros((index, 2), dtype = float)
readArray = np.zeros((read, 2), dtype = float)
count = 0

with open (file1, "r") as fh:
    for line in fh:
        line = line.strip("\n")
        line = line.split()
        readArray[count][0] = float(line[0])
        readArray[count][1] = float(line[1])
        count += 1
     
plt.bar(readArray[:, 0] + 1, readArray[:, 1])
plt.xlabel("# base pair")
plt.ylabel("Mean Quality Score")
plt.title("R1 Mean Quality Read Scores")
plt.show()

count = 0

with open (file2, "r") as fh:
    for line in fh:
        line = line.strip("\n")
        line = line.split()
        indexArray[count][0] = float(line[0])
        indexArray[count][1] = float(line[1])
        count += 1
     
plt.bar(indexArray[:, 0] + 1, indexArray[:, 1])
plt.xlabel("# base pair")
plt.ylabel("Mean Quality Score")
plt.title("R2 Mean Quality Index Scores")
plt.show()

indexArray = np.zeros((index, 2), dtype = float)
readArray = np.zeros((read, 2), dtype = float)
count = 0

with open (file3, "r") as fh:
    for line in fh:
        line = line.strip("\n")
        line = line.split()
        indexArray[count][0] = float(line[0])
        indexArray[count][1] = float(line[1])
        count += 1
     
plt.bar(indexArray[:, 0] + 1, indexArray[:, 1])
plt.xlabel("# base pair")
plt.ylabel("Mean Quality Score")
plt.title("R3 Mean Quality Index Scores")
plt.show()

count = 0

with open (file4, "r") as fh:
    for line in fh:
        line = line.strip("\n")
        line = line.split()
        readArray[count][0] = float(line[0])
        readArray[count][1] = float(line[1])
        count += 1
     
plt.bar(readArray[:, 0] + 1, readArray[:, 1])
plt.xlabel("# base pair")
plt.ylabel("Mean Quality Score")
plt.title("R4 Mean Quality Read Scores")
plt.show()
```
